```python
import pandas as pd
import numpy as np
```

```python
academico = pd.read_csv('/Educacion_Secundaria_Ecu.csv')
academico.head(4)
```

```python
## Conversion de Datos Numericos a Categoricos Ordinales
academico['Calificaion Redaccion Textos-Categorico'] = academico['Calificaion Redaccion Textos/100'].apply(lambda x: "Sobresaliente" 
                                                  if x >= 90 else "Satisfactorio" if x>=80 and x < 90 else "Muy Bueno" if 
                                                   x<80 and x>=70 else "Bueno" if x<70 and x>=60 else "Insuficiente")
academico.head(4)
```

```python
##Agregacion de Datos.
academico = academico.assign(Calificaciones_Totales = lambda x: x["Calificacion Matematicas/100"] + x["Calificacion Lectura Comprensiva/100"] 
                             + x["Calificaion Redaccion Textos/100"])
academico.head(4)
```

```python
## Filtrar Datos
Filtrados = academico[academico.apply((lambda x: (x['Calificaciones_Totales']/300*100)>80), axis=1)]
Filtrados.head(4)
```

```python
## Encodificacion Binaria
academico['Tipo Desayuno Binario'] = academico['Tipo Desayuno'].map(lambda x: 1 if x  =='Estandar' else 0)
academico.head(4)
```

```python
## Encodificacion categorica
academico_1 = academico.assign(Raza_Codificacion_Numerica = lambda x: (pd.factorize(academico['Raza'])[0]))
academico_1.iloc[:, -3:]
```

```python
## Estandardizacion
academico_1['Calificacion Matematicas Normalizado'] = academico_1.apply(lambda x: round((x['Calificacion Matematicas/100'] - 
        academico_1['Calificacion Matematicas/100'].mean()) / academico_1['Calificacion Matematicas/100'].std(), 2), axis=1)
academico_1.iloc[:, -3:]
```

```python
## Normalizacion Min-Max
academico_1['Calificacion Lectura MinMax'] = academico_1.apply(lambda x: round((x['Calificacion Lectura Comprensiva/100']
                                    -academico_1['Calificacion Lectura Comprensiva/100'].min())/
                                    (academico_1['Calificacion Lectura Comprensiva/100'].max()-
                                     academico_1['Calificacion Lectura Comprensiva/100'].min()),2), axis=1)
academico_1.iloc[:, -3:]
```

```python
## Rango Intercuantil y Valores Anomalos (Outliers)

Q1 = academico_1['Calificaion Redaccion Textos/100'].quantile(.25)
Q3 = academico_1['Calificaion Redaccion Textos/100'].quantile(.75)
IQR = Q3-Q1

academico_1['Calificaion Redaccion Textos Anomalos'] = academico_1['Calificaion Redaccion Textos/100'].\
    apply(lambda x: 'Valor Anomalo' if (x > Q3+1.5*IQR) or (x < Q1-1.5*IQR) else "Valor Normal")
academico_1.iloc[:, -3:]
```

```python
## Filtrado por Valor Anomal
anomales=academico_1[academico_1['Calificaion Redaccion Textos Anomalos']=='Valor Anomalo']
anomales[['Raza_Codificacion_Numerica', 'Calificacion Matematicas Normalizado', 'Calificaion Redaccion Textos Anomalos']]
```
